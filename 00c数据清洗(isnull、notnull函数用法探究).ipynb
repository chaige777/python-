### isnull()函数

```python
import pandas as pd
import numpy as np
df = pd.DataFrame(np.random.randint(10,99,size=(10,5)))
df.iloc[4:6,0] = np.nan
df.iloc[5:7,2] = np.nan
df.iloc[7,3] = np.nan
df.iloc[2:3,4] = np.nan
df
```

```python
df.isnull()
```

### 判断哪些列包含缺失值

```python
df.isnull().any()
```

### 反应每列缺失值的数量

```python
df.isnull().sum()
```

### notnull()函数

```python
df.notnull()
```
